```python
timefmt = "%d/%m/%y %S:%M:%H" 
stock['Date'] = pd.to_datetime(stock['Date'], format = timefmt, errors='coerce')
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(stock, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Applying scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = [ 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_test[num_vars] = scaler.fit_transform(df_test[num_vars])

df_train
```

```python
y_train = df_train.pop('Volume')
X_train = df_train
y_test = df_test.pop('Volume')
X_test = df_test
```
